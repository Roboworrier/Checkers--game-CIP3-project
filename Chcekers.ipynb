```python
#LET'S MAKE THE CHECKERS BOARD FIRST #

board = [
    [' ', 'O', ' ', 'O', ' ', 'O', ' ', 'O'],
    ['O', ' ', 'O', ' ', 'O', ' ', 'O', ' '],
    [' ', 'O', ' ', 'O', ' ', 'O', ' ', 'O'],
    [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' '],
    [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' '],
    ['X', ' ', 'X', ' ', 'X', ' ', 'X', ' '],
    [' ', 'X', ' ', 'X', ' ', 'X', ' ', 'X'],
    ['X', ' ', 'X', ' ', 'X', ' ', 'X', ' ']
]


def print_board(board):
    print('   0  1  2  3  4  5  6  7')
    print('  ------------------------')
    for i in range(len(board)):
        print(i, '|', end='')
        for j in range(len(board[i])):
            print(board[i][j], '|', end='')
        print('\n  ------------------------')

# Moving Positions of Pieces on Board

def get_valid_moves(board, player, row, col):
    valid_moves = []
    if player == 'X' or board[row][col] == 'XK':
        if row > 0:
            if col > 0 and board[row - 1][col - 1] == ' ':
                valid_moves.append((row - 1, col - 1))
            if col < 7 and board[row - 1][col + 1] == ' ':
                valid_moves.append((row - 1, col + 1))
        if board[row][col] == 'XK' and row < 7:
            if col > 0 and board[row + 1][col - 1] == ' ':
                valid_moves.append((row + 1, col - 1))
            if col < 7 and board[row + 1][col + 1] == ' ':
                valid_moves.append((row + 1, col + 1))
    if player == 'O' or board[row][col] == 'OK':
        if row < 7:
            if col > 0 and board[row + 1][col - 1] == ' ':
                valid_moves.append((row + 1, col - 1))
            if col < 7 and board[row + 1][col + 1] == ' ':
                valid_moves.append((row + 1, col + 1))
        if board[row][col] == 'OK' and row > 0:
            if col > 0 and board[row - 1][col - 1] == ' ':
                valid_moves.append((row - 1, col - 1))
            if col < 7 and board[row - 1][col + 1] == ' ':
                valid_moves.append((row - 1, col + 1))
    return valid_moves

# Defining moving positions of Player1 = X, Player2 = O

def make_move(board, player, row, col, move_row, move_col):
    board[move_row][move_col] = board[row][col]
    board[row][col] = ' '
    if move_row == 0 and board[move_row][move_col] == 'X':
        board[move_row][move_col] = 'XK'
    if move_row == 7 and board[move_row][move_col] == 'O':
        board[move_row][move_col] = 'OK'


def is_game_over(board):
    x_count = 0
    o_count = 0
    for i in range(len(board)):
        for j in range(len(board[i])):
            if board[i][j] == 'X' or board[i][j] == 'XK':
                x_count += 1
            elif board[i][j] == 'O' or board[i][j] == 'OK':
                o_count += 1
    return x_count == 0 or o_count == 0

#Starting and Ending Process Of The Game

def main():
    print("Welcome to Checkers!")
    print("X plays first. O plays second.")
    player = 'X'
    while not is_game_over(board):
        print_board(board)
        valid_move = False
        while not valid_move:
            if player == 'X':
                row = int(input("X, enter the row of the piece you want to move: "))
                col = int(input("X, enter the column of the piece you want to move: "))
            else:
                row = int(input("O, enter the row of the piece you want to move: "))
                col = int(input("O, enter the column of the piece you want to move: "))
            valid_moves = get_valid_moves(board, player, row, col)
            if valid_moves:
                print("Valid moves:")
                for i, move in enumerate(valid_moves):
                    print(f"{i + 1}: {move}")
                move_num = int(input("Enter the number of the move you want to make: ")) - 1
                move_row, move_col = valid_moves[move_num]
                make_move(board, player, row, col, move_row, move_col)
                valid_move = True
            else:
                print("Invalid move. Please try again.")
        if player == 'X':
            player = 'O'
        else:
            player = 'X'
    print_board(board)
    if player == 'X':
        print("O wins!")
    else:
        print("X wins!")


if __name__ == '__main__':
    main()
```
